### Q 1.12 - <span style="color:red">[15]</span> - Calculate confidence interval (based on 99% confidence level) for the mean of target by bootstrapping. For this purpose, code a bootstrap function that in each bootstrap iteration, samples from the training set to fit the linear regression model and uses the test set to make predictions - therefore your bootstrap statistic is the average of the predictions over the test set. Your function must take as input arguments: your model, Xtrain, ytrain, Xtest, and numboot=100. The function must return only one object that is the array of recorded values for the bootstrap statistic in $\mathrm{euros}$ - and not $log(\mathrm{euros})$. Also, the unit of the confidence interval must be $\mathrm{euros}$ - and not $log(\mathrm{euros})$.

```python
#

def bootstraps(model, X_train, X_test, y_train, y_test, numboot=100):
    
    n = len(X_train)
    
    means = np.zeros(numboot)
    
    for i in range(numboot):
        
        X_training_sample = X_train.sample(n, replace=True) # Sample the training set
        y_training_sample = y_train.sample(n, replace=True)
        
        fit = model.fit(X_training_sample, y_training_sample) # Fit to the sample
        
        pred = model.predict(X_test) # Predict on the test set
        
        means[i] = pred.mean() # Get the mean
        
    return means

X = df.drop(['Value', 'Log Value'], axis=1)
y = df['Value']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=seed, test_size=0.2)

model = LinearRegression()

bstrap_means = bootstraps(model, X_train, X_test, y_train, y_test) # Generate bootstrap means in euros.

# The following code to generate the CI was adapted from Lab 4.

bm = pd.DataFrame(data=bstrap_means-df.Value.mean(), columns=['Value'])

boot_CL = 99/100 
p_1 = (1-boot_CL)/2
p_2 = 1-p_1
boot_quant = np.quantile(bm, [p_1, p_2])

boot_ci = [(df.Value.mean() - boot_quant[1]).round(3), 
           (df.Value.mean() - boot_quant[0]).round(3)]
print("Boot Confidence Interval:",boot_ci)
        
```

### Q 1.13 - <span style="color:red">[6]</span> - Construct a 99% confidence interval using the Central Limit Theorem (again in $\mathrm{euros}$). 

```python
# Again, this is mostly adapted from Lab 4 - except, we use the critical value 2.576 for 99% probability.

stderr = np.std(df.Value) / np.sqrt(len(df.Value))

critval = 2.576 # critical value

norm_ci = [(df.Value.mean() - critval*stderr).round(3), 
           (df.Value.mean() + critval*stderr).round(3)]

print("Norm ci:",norm_ci)
```

### Q 1.14 - <span style="color:red">[10]</span> - We want to see the effect of sample size ($n$) on the CI calculated from CLT. Write a `for` loop which in each iteration randomly samples from your "sample statistic" and calculates and stores the width of the corresponding CI in an array. Obviously, you should start from a small $n$ and increase it per iteration. After the loop, plot sample size (*i.e.*, $n$) against CI width and report your observation in one sentence.

```python
#

widths = []

for i in range(100, 10000):
    sample = df['Value'].sample(i, replace=True)
    
    stderr = np.std(sample) / np.sqrt(len(sample))

    critval = 2.576 # critical value
    
    # Find CI
    norm_ci = [(sample.mean() - critval*stderr).round(3), 
               (sample.mean() + critval*stderr).round(3)]
        
    width = norm_ci[1] - norm_ci[0]
    
    widths.append(width)
    
x_grid = np.linspace(100, 10000, 9900)

plt.scatter(x_grid, widths, alpha=0.3)

plt.xlabel("Sample Size")
plt.ylabel("Width of CI for Value (millions of euros)")

plt.show()
```

#### *Answer*

One sees that as $n$ increases, the width decreases, as a larger sample size gives higher precision.

### Q 1.15 - <span style="color:red">[9]</span> - Randomly subsample your "sample statistic" with $n=30$ and calculate $t$-based 99% CI (in $\mathrm{euros}$). Is it a good idea to calculate CI for this data set this way? Why?

Hint: It would be a good idea to run your code for this part a few times prior to answering the question.

```python
# Adapted from Lab 4.

alpha = (1/100)
crit_val = 1-(alpha/2)

dof = 29 # n-1 = 30-1 = 29

t_val = t.ppf(crit_val, df=dof)

sample = df['Value'].sample(30, replace=True)
    
stderr = np.std(sample) / len(sample)

t_ci = np.mean(sample) + t_val * stderr * np.array([-1, 1])

print('The t-based confidence interval is equal to {}'.format(t_ci.round(3)))
```

#### *Answer*

It is not a good idea to sample this way. Setting $n$ = 30 is a low sample size, so this will introduce a high amount of variance in each sample, which will in turn introduce a high amount of variance in the confidence interval. We can see this by running the code several times. Some of the CIS are reasonable, whereas some of them are completely off. If we increase $n$, then our CIs will be more precise. But with $n = 30$, this is not a good method. Bootstrapping is better.
